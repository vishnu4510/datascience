```python
import pandas as pd

df = pd.read_csv('/content/Software Engineer Salaries.csv')
```

 **Q1: What is the size of the dataset? (Number of rows and columns)**

```python

num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

**Q2: What are the data types of each column?**

```python

data_types = df.dtypes
print(data_types)
```

**Q3: Are there any missing values in the dataset? If so, which columns have missing values and how many?**

```python

missing_values = df.isnull().sum()


missing_values = missing_values[missing_values > 0]
print(missing_values)
```

**Q4: What are the unique values in a categorical column of your choice? How many unique values are there?**

```python

categorical_column = 'Company'


unique_values = df[categorical_column].unique()
num_unique_values = len(unique_values)

print(f"Unique values in '{categorical_column}': {unique_values}")
print(f"Number of unique values: {num_unique_values}")
```

**Q5: For a numerical column of your choice, calculate the mean, median, and standard deviation.**

```python

numerical_column = 'Company Score'

mean = df[numerical_column].mean()
median = df[numerical_column].median()
std_dev = df[numerical_column].std()

print(f"Mean of '{numerical_column}': {mean}")
print(f"Median of '{numerical_column}': {median}")
print(f"Standard Deviation of '{numerical_column}': {std_dev}")
```

**Q6: Identify any potential outliers in a numerical column of your choice. Explain your approach.**

```python

numerical_column = 'Company Score'

Q1 = df[numerical_column].quantile(0.25)
Q3 = df[numerical_column].quantile(0.75)


IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = df[(df[numerical_column] < lower_bound) | (df[numerical_column] > upper_bound)]

print(f"Outliers in '{numerical_column}':")
print(outliers)
```

**Q7: Find the correlation between numerical columns and discuss any interesting relationships.**

```python
# Convert non-numeric columns to numeric if applicable (errors='coerce' converts non-convertible values to NaN)
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
df_numeric = df_numeric.dropna()

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

print("Correlation matrix:")
print(correlation_matrix)
```

**Q8: Group the data by a categorical column and compute the aggregate statistics (e.g., mean, sum) for a numerical column.**

```python

categorical_column = 'Company'
numerical_column = 'Company Score'

grouped_stats = df.groupby(categorical_column)[numerical_column].agg(['mean', 'sum', 'count'])

print(f"Aggregate statistics for '{numerical_column}' grouped by '{categorical_column}':")
print(grouped_stats)
```
