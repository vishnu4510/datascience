**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

**Diabetes Dataset EDA:
Univariate Analysis:
Analyze individual variables (e.g., histograms, box plots)**

````python
import pandas as pd

diabetes_df = pd.read_csv('/content/diabetes_model.csv')

print(diabetes_df.head())
````

````python
#hist
diabetes_df.hist(figsize=(10, 10))
````

`*diabetes_df.hist()` creates histograms for all numerical columns.
`figsize=(10, 10)` sets the size of the plot to 10 inches wide and 10 inches tall.

This helps visualize the distribution of data in each column.
 You can identify patterns like skewness, outliers, and the general shape of the data.

````python
diabetes_df.boxplot(figsize=(10, 10))
````

**Bivariate Analysis:
Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices)**



````python


# Scatter plots
plt.figure(figsize=(10, 6))
plt.scatter(diabetes_df['Pregnancies'], diabetes_df['Glucose'],color='Green')
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.title('Pregnancies vs Glucose')
plt.show()

# correlation matrices
corr_matrix = diabetes_df.corr()
plt.figure(figsize=(10, 8))
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
````

The scatter plot shows the relationship between pregnancies and glucose levels, helping to identify any patterns or outliers. The correlation matrix heatmap reveals the strength and direction of relationships between all numerical features, highlighting potential multicollinearity and important feature associations.

**Multivariate Analysis:
Investigate interactions among multiple variables (e.g., pair plots) any  three only**

````python


import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(diabetes_df[['Pregnancies', 'Glucose', 'BloodPressure', 'Outcome']], hue='Outcome',palette='coolwarm',markers='+')

plt.show()
````

 It shows the distribution of each variable (on the diagonal) and the scatter plots between pairs of variables.
 - By using 'Outcome' as the hue, you can see how the distribution and relationships differ between patients with and without diabetes.
 - This visualization can help identify potential patterns or correlations between variables and the outcome of diabetes.
 - For example, you might observe that higher glucose levels are more common in patients with diabetes, or that there's a relationship between pregnancies and blood pressure in diabetic patients.
